```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from random import gauss, randrange, random
```

```python
import statsmodels.graphics.tsaplots as sm
import statsmodels.api as smt
```

```python
from statsmodels.tsa.arima_process import ArmaProcess
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
def acf_pacf(x, qtd_lag):
    fig = plt.figure(figsize=(16,10))
    ax1 = fig.add_subplot(221)
    #plot acf
    fig = sm.plot_acf(x, lags=qtd_lag, ax=ax1)
    ax2 = fig.add_subplot(222)
    #plot pacf
    fig = sm.plot_pacf(x, lags=qtd_lag, ax=ax2)
    plt.show()
```

### Considerações:

AR e MA incluem o componente lag zero. E geralmente ele deve ser = 1 

Por conta da representação, os parâmetros para o AR devem ter sinais diferentes da representação original do ARIMA


#### Simulando um processo AR(1)

$$
x_t = -0.6 x_{t-1} + z_t
$$

```python

ar1 = np.array([1, 0.6])
ma1 = np.array([1])
AR_object1 = ArmaProcess(ar1, ma1)
rs_ar1 = AR_object1.generate_sample(nsample=1000)

fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(111)
ax.plot(rs_ar1);
```

```python
AR_object1.isinvertible
```

```python
AR_object1.isstationary
```

```python
plt.show()
```

```python
acf_pacf(rs_ar1, 20)
```

Para um AR(1) em que  $\phi_1$ negativo, as correlações vão decaindo até 0 de acordo com o aumento do lag, porém existe uma alteração entre os sinais dos valores
